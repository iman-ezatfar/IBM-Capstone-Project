**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
QUERY = """
SELECT CompTotal
FROM main
WHERE CompTotal IS NOT NULL AND CompTotal < 300000
"""
comp_df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(10, 6))
sns.histplot(data=comp_df, x='CompTotal', bins=30, kde=True, color='black')
plt.title('Distribution of Total Compensation (CompTotal)')
plt.xlabel('Total Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
QUERY = """
SELECT DISTINCT AGE
FROM main
"""
age_df = pd.read_sql_query(QUERY, conn)
age_df
```

```python
QUERY = """
SELECT Age
FROM main
WHERE Age IS NOT NULL AND Age != 'Prefer not to say'
"""

age_df = pd.read_sql_query(QUERY, conn)

age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 29,
    '35-44 years old': 39,
    '45-54 years old': 49,
    '55-64 years old': 59,
    '65 years or older': 70
}

age_df['AgeNumeric'] = age_df['Age'].map(age_map)
age_df['AgeNumeric'].value_counts()

plt.figure(figsize=(8, 6))
sns.boxplot(x=age_df['AgeNumeric'], color='cornflowerblue')
plt.title('Box Plot of Age')
plt.xlabel('Age (Years)')
plt.tight_layout()
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
QUERY = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL AND WorkExp IS NOT NULL AND Age != 'Prefer not to say'
"""
df_age_exp = pd.read_sql_query(QUERY, conn)
df_age_exp['AgeNumeric'] = df_age_exp['Age'].map(age_map)
df_age_exp['WorkExp'] = pd.to_numeric(df_age_exp['WorkExp'], errors='coerce')
df_age_exp = df_age_exp.dropna(subset=['AgeNumeric', 'WorkExp'])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_age_exp, x='AgeNumeric', y='WorkExp', alpha=0.5)
plt.title('Scatter Plot of Age vs Work Experience')
plt.xlabel('Age (Years)')
plt.ylabel('Work Experience (Years)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
query = """
SELECT DatabaseWantToWorkWith, COUNT(*) as Count
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL
GROUP BY DatabaseWantToWorkWith
ORDER BY Count DESC
LIMIT 5
"""

df_db = pd.read_sql_query(query, conn)

plt.figure(figsize=(8,8))
plt.pie(
    df_db['Count'],
    labels=df_db['DatabaseWantToWorkWith'],
    autopct='%1.1f%%',
    startangle=140,
    shadow=False
)
plt.title("Top 5 Databases Respondents Want to Learn Next Year")
plt.axis('equal') 
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
query = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IN ('25-34 years old', '35-44 years old')
  AND TimeSearching IS NOT NULL
  AND TimeAnswering IS NOT NULL
"""

df_age = pd.read_sql_query(query, conn)
df_age
```

```python
query = """
SELECT Age, TimeSearching, TimeAnswering
FROM main
WHERE Age IN ('25-34 years old', '35-44 years old')
  AND TimeSearching IS NOT NULL
  AND TimeAnswering IS NOT NULL
"""

df_age = pd.read_sql_query(query, conn)

time_map = {
    'Less than 15 minutes a day': 7.5,
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day': 45,
    '60-120 minutes a day': 90,
    'More than 120 minutes a day': 150
}

df_age['TimeSearchingNum'] = df_age['TimeSearching'].map(time_map)
df_age['TimeAnsweringNum'] = df_age['TimeAnswering'].map(time_map)

median_times = df_age.groupby('Age')[['TimeSearchingNum', 'TimeAnsweringNum']].median()

median_times.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Set1')

plt.title('Median Time Searching and Time Answering by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Time (minutes)')
plt.xticks(rotation=0)
plt.legend(['Time Searching', 'Time Answering'])
plt.tight_layout()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
query = """
SELECT Age, CompTotal
FROM main
WHERE Age IN ('45-54 years old', '55-64 years old')
  AND CompTotal IS NOT NULL
"""

df_age_comp = pd.read_sql_query(query, conn)

df_age_comp['CompTotal'] = pd.to_numeric(df_age_comp['CompTotal'], errors='coerce')

df_age_comp = df_age_comp.dropna(subset=['CompTotal'])

median_comp = df_age_comp.groupby('Age')['CompTotal'].median().sort_index()

median_comp.plot(kind='bar', figsize=(8,6), color='Gray')

plt.title('Median Total Compensation by Age Group (45-60 years)')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (CompTotal)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
query = """
SELECT MainBranch, COUNT(*) AS Count
FROM main
GROUP BY MainBranch
ORDER BY Count DESC
"""

df_mainbranch = pd.read_sql_query(query, conn)

# Plot horizontal bar chart
plt.figure(figsize=(18,6))
plt.barh(df_mainbranch['MainBranch'], df_mainbranch['Count'], color='teal')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Distribution of MainBranch')
plt.gca().invert_yaxis()  # largest bar on top
plt.tight_layout()
plt.show()
```
