### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
df.head()
```

<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
summary = pd.DataFrame({
    'Data Type': df.dtypes,
    'Non-Null Count': df.notnull().sum(),
    'Missing Value' : df.isnull().sum(),
    })
summary
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
df['Country'].unique()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
df['Country'] = df['Country'].str.strip()
df['Country'] = df['Country'].str.title()
df['Country'].fillna('Unknown', inplace=True)

df['Country'] = df['Country'].replace({
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'Russian Federation': 'Russia',
    'Republic of Korea': 'South Korea',
    'Iran, Islamic Republic of...': 'Iran',
    'Venezuela, Bolivarian Republic of...': 'Venezuela',
    # Add more as needed
})
df['Country'].unique()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
employment_dummies = pd.get_dummies(df['Employment'], prefix='Employment')
employment_dummies = employment_dummies.applymap(lambda x: 1 if x else 0)


#OPTION1
df_encoded = pd.concat([df, employment_dummies], axis=1)
df_encoded.head()

#OPTION2
#df.drop('Employment', axis=1, inplace=True)
#df = pd.concat([df, employment_dummies], axis=1)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
df_encoded.isna().sum().sort_values(ascending=False).head(1)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
mean_CompYearly = df_encoded['ConvertedCompYearly'].mean()
df_encoded['ConvertedCompYearly'].fillna(mean_CompYearly, inplace=True)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
freq_remote_work = df['RemoteWork'].value_counts().idxmax()
df_encoded['RemoteWork'] = df_encoded['RemoteWork'].fillna(freq_remote_work)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
norm_CompYear = (df_encoded['ConvertedCompYearly'] - df_encoded['ConvertedCompYearly'].min()) / (df_encoded['ConvertedCompYearly'].max() - df_encoded['ConvertedCompYearly'].min())
df_encoded['ConvertedCompYearly'] = norm_CompYear
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np

df_encoded['ConvertedCompYearly'] = np.log(df_encoded['ConvertedCompYearly']+1)
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df_encoded['YearsCodePro'].astype
```

```python
def categorize_experience(years):
    if isinstance(years, str):
        if years == 'Less than 1 year':
            years = 0
        elif years == 'More than 50 years':
            years = 51
        else:
            try:
                years = float(years)
            except ValueError:
                return 'Unknown'
    elif pd.isnull(years):
        return 'Unknown'

    if years < 2:
        return 'Beginner'
    elif years < 5:
        return 'Intermediate'
    else:
        return 'Expert'
    
df_encoded['ExperienceLevel'] = df_encoded['YearsCodePro'].apply(categorize_experience)
df_encoded['ExperienceLevel'].value_counts()
```
